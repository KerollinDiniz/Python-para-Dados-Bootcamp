10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python

df_saude['Passos diários'] = pd.to_numeric(df_saude['Passos diários'], errors='coerce')
print(df_saude['Passos diários'].isna().sum())

frequencia_cardiaca_acima_70 = df_saude[df_saude['Frequência cardíaca'] > 70]
frequencia_cardiaca_menor_igual_70 = df_saude[df_saude['Frequência cardíaca'] <= 70]

media_passos_acima_70 = frequencia_cardiaca_acima_70['Passos diários'].mean()
media_passos_menor_70 = frequencia_cardiaca_menor_igual_70['Passos diários'].mean()

print(f"Frquencia cardíaca acima de 70, Media de Passos:{media_passos_acima_70}")
print(f"Frquencia cardíaca menor ou igual a 70, Media de Passos: {media_passos_menor_70}")

if media_passos_acima_70 > media_passos_menor_70:
    print("As pessoas com frequência cardíaca acima de 70 dão mais passos.")
else:
    print("As pessoas com frequência cardíaca menor ou igual a 70 dão mais passos.")
```
